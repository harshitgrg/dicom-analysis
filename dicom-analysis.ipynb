```python
import pydicom as pyd
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
```

```python
dicom_path="./Lung-DICOM"
dcm_files=[]
for root,dirs,files in os.walk(dicom_path):
   for filename in files:# To seperate the files
    if ".dcm" in filename.lower():
        dcm_files.append(filename)

print(dcm_files[0])
dc=pyd.dcmread(dicom_path+"/"+dcm_files[0])# This is our single file
# Trying to understand Parameters like Pixel Spacing etc. I will use them later.
print(dc)
print(dc.Rows)# Trying to understand the elements of that dicom file.
print(dc.Columns)
print(len(dcm_files))
print(dc.pixel_array.shape)
plt.imshow(dc.pixel_array,cmap=plt.cm.bone)
```
